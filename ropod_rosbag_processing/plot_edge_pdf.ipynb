```python
import statistics
import os
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
def get_files(path):
    edge_files = dict()

    for root, directories, files in os.walk(path):
        if root != path:
            try:
                edge_name = root.replace(path, '')
                index = files.index('all_runs.out')
                edge_files[edge_name] = root + '/' + files[index]
            except ValueError:
                print("All runs file not present in", root)

    return edge_files
```

```python
def get_duration(file):
    durations = list()
    with open(file, 'r') as file:
        for line in file.readlines():
            duration = float(line.split(' ')[1])
            durations.append(duration)
    return durations
```

```python
def plot_pdfs(edge_files):
    for edge_name, output_file in edge_files.items():
        print("-----------------------------------------")
        print("Edge: ", edge_name)
        durations = get_duration(output_file)
        print("# entries: ", len(durations))

        if len(durations) > 1:
            mu = statistics.mean(durations)
            sigma = statistics.stdev(durations)
            x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
            print("mean: ", mu)
            print("standard deviation: ", sigma)

            plt.plot(x, stats.norm.pdf(x, mu, sigma))
            plt.grid()
            plt.show()
```

```python
edge_files = get_files('angela/')
plot_pdfs(edge_files)
```
